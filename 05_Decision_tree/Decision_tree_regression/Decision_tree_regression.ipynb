# Decision Tree Regression

## Introduction

Decision Tree Regression is a supervised machine learning algorithm used to predict continuous target values.  
It works by splitting the dataset into smaller subsets based on feature values and forming a tree-like structure of decisions.

Unlike linear models, Decision Trees can capture **non-linear relationships** between features and the target variable.

---

##  How Decision Tree Regression Works

1. The algorithm selects the best feature and split point.
2. The dataset is divided into subsets based on that split.
3. This process is repeated recursively to create branches.
4. Each leaf node contains the predicted value, usually the **mean of target values** in that region.

---

##  Splitting Criteria

Decision Tree Regression uses **variance reduction** to decide the best split.

Objective:

Minimize the variance of target values within each node.

Common metrics:
- Mean Squared Error (MSE)
- Mean Absolute Error (MAE)

---

##  Key Hyperparameters

- **max_depth** → Maximum depth of the tree
- **min_samples_split** → Minimum samples required to split a node
- **min_samples_leaf** → Minimum samples required in a leaf node
- **max_features** → Number of features considered for splitting

---

##  Advantages

- Handles non-linear relationships
- No need for feature scaling
- Easy to interpret and visualize
- Works with both numerical and categorical features

---

##  Limitations

- Prone to overfitting
- Sensitive to small variations in data
- Can create complex trees if not pruned

---

## Applications

- House price prediction
- Sales forecasting
- Demand prediction
- Risk assessment

---

### Import requiremnts 

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
from sklearn.preprocessing import StandardScaler
```

### Load Dataset 

```python
data=pd.read_csv('bangalore house price prediction.csv')
```

```python
data.head()
```

```python
data.shape
```

## Train test split 

```python
X=data.drop(['price'],axis=1)
y=data['price']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Model Train 

```python
model=DecisionTreeRegressor(criterion='absolute_error',max_depth=9)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

## Apply StandardScaler

```python
sc=StandardScaler()
```

```python
X_train_sc=sc.fit_transform(X_train)
X_test_sc=sc.transform(X_test)
```

```python
model2=DecisionTreeRegressor()
```

```python
model2.fit(X_train_sc,y_train)
```

```python
model2.score(X_test_sc,y_test)
```
