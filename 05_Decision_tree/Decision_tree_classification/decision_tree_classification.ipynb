# Decision Tree Classification

## Introduction

Decision Tree Classification is a supervised machine learning algorithm used to predict categorical outcomes.  
It works by recursively splitting the dataset based on feature values to create a tree-like structure of decisions.

Each internal node represents a decision based on a feature, each branch represents an outcome of that decision, and each leaf node represents the final predicted class.

---

##  How Decision Trees Work

1. Select the best feature to split the data.
2. Divide the dataset into subsets based on that feature.
3. Repeat the process recursively to build branches.
4. Continue until stopping criteria are met.
5. Leaf nodes represent the predicted class.

---

##  Splitting Criteria

Decision Trees use impurity measures to determine the best split.

Common criteria:

**1. Gini Impurity**

Gini = 1 - Σ (pᵢ²)

**2. Entropy (Information Gain)**

Entropy = - Σ pᵢ log₂(pᵢ)

The algorithm selects the split that **reduces impurity the most**.

---

## Key Hyperparameters

- **max_depth** → Maximum depth of the tree
- **min_samples_split** → Minimum samples required to split a node
- **min_samples_leaf** → Minimum samples required in a leaf node
- **criterion** → Measure used for splitting (gini or entropy)

---

##  Advantages

- Easy to understand and interpret
- Works with both numerical and categorical data
- No feature scaling required
- Handles non-linear relationships

---

##  Limitations

- Prone to overfitting
- Sensitive to small variations in data
- Large trees can become complex

---

##  Applications

- Medical diagnosis
- Customer segmentation
- Fraud detection
- Credit risk assessment


### Import Requirements 

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score
from matplotlib.pylab import rcParams 
from sklearn.tree import plot_tree
```

### load dataset 

```python
iris=load_iris()
```

```python
X=iris.data
y=iris.target
```

### Train test split 

```python
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Model Traning 

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

### Check Accuracy score 

```python
accuracy_score(y_test,y_pred)
```

### Ploting Tree 

```python
from matplotlib.pylab import rcParams 
rcParams['figure.figsize'] = 80,50

plot_tree(clf);
```
